
# Loops

```python
fruits=["Banana", "Mango","Apple"]
for x in fruits:
 print(x)
```

```python
fruits=["Banana", "Mango", "Apple"]
for x in fruits:
 print(x)
 if x == "Banana":
  break
```

```python
fruits=["Mango", "Banana", "Apple"]
for x in fruits:
 print(x)
 if x== "Banana":
  continue
```

```python
for x in range(2):
 print(x)
```

```python
for x in range(1,11):
 print(x)
```
