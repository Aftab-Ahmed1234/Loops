# For Loop

```python
fruits=["Banana", "Mango","Apple"]
for x in fruits:
 print(x)
```

```python
fruits=["Banana", "Mango", "Apple"]
for x in fruits:
 print(x)
 if x == "Banana":
  break
```

```python
fruits=["Mango", "Banana", "Apple"]
for x in fruits:
 print(x)
 if x== "Banana":
  continue
```
